# Take

- First argument : indices

- indices will take some values in array : in a list

- It tries to extract the data

- That values we provide represents columns or rows

- Axis =0 is rows

- Axis = 1 is columns

- By default axis mentioned as zero

```python
visa_df.take([100,200,300])
# IT retrives 3 rows which  having index 100,200,300
```

```python
visa_df.take([10,11],axis=1) # axis = 1 is columns  index start = 0 
```

```python
visa_df.take([1,4,6],axis=1)
```

```python
# case-3: select 3 columns 2 ,5,and 7 
# using take 
visa_df.take([2,5,7],axis=1)
```

```python
# case-4 slicing will fail
visa_df.take([2:5])
```

```python
# 100, 200 ,300  rows from # 2, 5,7 columns
```
